### Self_Number
* d(n) = sum(n의자리수들) + n = generator. self number = not generator. for example 1, 3, 5, 7, 9, 20, 31...
* 1 to less than 5000. sum of self number???

```python
def d(num):
    str_num = str(num)
    sum = 0
    for i in str_num:
        sum += int(i)
    return sum + num
```

```python
d(91)
```

```python
result_G = []

for i in range(1, 5000):
    result_G.append(d(i))
    G = result_G
    
print(G)
```

```python
result_S = []

for j in range(1, 5000):
    if j not in G:
        result_S.append(j)
        
sum(result_S)
```

### Notice_paging
* Return all pages by entering the total number of posts and the number of posts.

```python
def paging(m, n):
    page = m // n
    if m % n > 0:
        page += 1
    return page
```

```python
paging(1, 10)
```

```python
import pandas as pd

data = pd.DataFrame(columns={'m', 'n'}, data=[[0, 1], [1, 1], [2, 1], [1, 10], [10, 10], [11, 10]])
data
```

```python
result = []

for i in range(6):
    data_paging = paging(data.loc[i]['m'], data.loc[i]['n'])
    result.append(data_paging)
    
print(result)
data['result'] = result
data
```

### Distance_pair
* Given a one-dimensional point, write a function that outputs a pair of the smallest of them.(Assume that all the arrays of endpoints are aligned.)

```python
import numpy as np

points = np.random.randint(1, 1000, size=10)
points.sort()
points
```

```python
ds = []

for i in range(len(points)-1):
    first = points[i]
    second = points[i+1]
    
    distance = abs(first - second)

    ds.append(distance)
    
    if min(ds) == distance:
        d = [first, second]
        
print(points)
print(ds)
print(d)
```

### Change Tab to Space
* Write program that change Tab to 4 space
----------------------------------------
* with open에서 파일의 인코딩을 확인

```python
import os
import re

filepath = input('Enther path where file is located : ')
ext = input("Enter the extension : ")
```

```python
_filelist = []

for cur, dirs, files in os.walk(filepath):
    for filenm in files:
        if (filenm.lower()).endswith(ext.lower()):
            _filelist.append(cur + "\\" + filenm)
    
print(_filelist)
```

```python
for _path in _filelist:
    print(_path)
    os.system("copy {0} {1}".format(_path, _path+".txt"))
```

```python
for _path in _filelist:
    with open(_path+".txt", 'r', encoding='UTF-8') as f:
        srcContent = f.read()
        with open(_path, 'w', encoding='UTF-8') as writelist:
            writelist.write(srcContent.replace("\t", "    "))
```

### Perfect_num

```python
def d_ftn(N):
    l = []

    for i in range(1, N):
        if N % i == 0:
            l.append(int(i))

    if sum(l) == N:
        number=N
        print(N)
    else:
        print('Not perfect number')
```
